# Beautiful Soup

There are four main object types in Beautiful Soup:
- **BeautifulSoup object**
    - is a representation of the document scraped as a whole. It's easily navigable and searchable.
- **Tag object**
    - correspond to XML and HTML elements in an original document.
    - Reference data can navigated using tag attributes.  
- **NavigableString object**
    - used to add a bit of text within tag. 
    - Beautiful Soup uses NavigableString class as a container for bits of text. 
- **Common object**
    - is a type of NavigableString object that can be used for commenting code. 

```python
content = response.content
soup = BeautifulSoup(content ,'html.parser')
```

## Data Parsing

Methods for searching and filtering a parse tree using *find_all()*:
- Name argument
- Keyword argument
- String argument
- Lists
- Boolean argument
- Strings
- Regular expressions

```python
text_soup = soup.get_text()

text_soup
```

#### Name argument

```python
soup.find_all('a')
```

```python
soup.find_all('header')
```

#### Keyword argument

```python
soup.find_all(style='display: none;')
```

#### String arguments

```python
soup.find_all(['li', 'a'])
```

#### Regular Expressions

```python
regex = re.compile('t')

for result in soup.find_all(regex):
    print(result.name)
```

```python
soup.find_all(string = re.compile('You'))
```

#### Boolean

```python
for result in soup.find_all(False):
    print(result.name)
```
