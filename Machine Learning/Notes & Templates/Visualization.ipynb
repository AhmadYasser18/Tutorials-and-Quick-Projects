```python
import numpy as np
import pandas as pd

from pandas import DataFrame, Series

import matplotlib.pyplot as plt

import seaborn as sns 
```

Types of data visualization:

- **Data storytelling**: for generating presentations that would be used to present to organizational decision makers. 
    - Purpose: make it easy for the audience to understand the point to be made with the data visualization.
    - Visualization needs to be *clutter-free* and *highly focused*.
    - Audience: non analysts and non-technical business managers, decision makers and leaders

- **Data showcasing**: for generating presentations for analysts, scientists, mathematicians, and engineers. 
    - Purpose: showcase a lot of data so that conclusions could be drawn.
    - Visualization is highly contextual and open-ended
    - Audience: analysts, quants, engineers, mathematicians, scientists

- **Data art**: for presentations when presenting to activists or the general public at large.
    - Purpose: used in something like data journalism, or if you're trying to get your community or a community of activists to be engaged in taking action, then you would benefit by generating a piece of art that's based on data. Data art is a tool to educate, inform, and motivate the public in most cases.
    - Visualization should be attention-getting, creative, controversial.
    - Audience: idealists, dreamers, artists, and social activists.  

AI tool for color selection: [Colormind](http://colormind.io/)

```python
data = DataFrame({'Names': ['PersonA', 'PersonB', 'PersonC', 'PersonD', 'PersonE', 'PersonF', 'PersonG'],
                  'Age': np.random.randint(20, 30,7),
                  'Gender': [['M','F'][np.random.randint(2)] for i in range(7)],
                  'Rank': np.random.randint(1, 8, 7)})

data
```

## Bar Chart

```python
plt.bar(data['Names'], data['Age'])

plt.xlabel('Names')
plt.ylabel('Age')

plt.title('People Ages')

plt.show()
```

```python
plot = sns.barplot(data = data, x='Names', y='Age')
plot.set_title("People's Ages")
plt.show()
```

## Line Plot

```python
plt.plot(data['Names'], data['Age'])

plt.xlabel('Names')
plt.ylabel('Age')

plt.title('People Ages')

plt.show()
```

```python
plot = sns.lineplot(data = data, x='Names', y='Age')
plot.set_title("People's Ages")
plt.show()
```

## Pie Chart

```python
gender_count = data['Gender'].value_counts()
gender_count
```

```python
plt.pie(gender_count, labels=gender_count.index)

plt.title('Gender')

plt.show()
```

```python
colors = sns.color_palette('pastel')[:5]

plt.pie(gender_count, labels=gender_count.index, colors = colors)
```
