Types of data visualization:

- **Data storytelling**: for generating presentations that would be used to present to organizational decision makers. 
    - Purpose: make it easy for the audience to understand the point to be made with the data visualization.
    - Visualization needs to be *clutter-free* and *highly focused*.
    - Audience: non analysts and non-technical business managers, decision makers and leaders

- **Data showcasing**: for generating presentations for analysts, scientists, mathematicians, and engineers. 
    - Purpose: showcase a lot of data so that conclusions could be drawn.
    - Visualization is highly contextual and open-ended
    - Audience: analysts, quants, engineers, mathematicians, scientists

- **Data art**: for presentations when presenting to activists or the general public at large.
    - Purpose: used in something like data journalism, or if you're trying to get your community or a community of activists to be engaged in taking action, then you would benefit by generating a piece of art that's based on data. Data art is a tool to educate, inform, and motivate the public in most cases.
    - Visualization should be attention-getting, creative, controversial.
    - Audience: idealists, dreamers, artists, and social activists.

## Object Oriented PLotting

Adding plot elements is an essential part of Object-oriented plotting. A plot is built by first generating a blank figure object and then populating that object with plots and plot elements.   
There are four steps in Object-oriented plotting:  
- Create a blank figure object
- Add axes to the figure
- Generate a plot or plots within the figure object.
- Specify plotting and layout parameters for the plots within the figure.

One last thing to touch on here is subplots, a subplot is a plotting figure that contains more than one plot or subplots. It's easy to generate a subplot in Matplotlib with the subplots function.

```python
%matplotlib inline 
#tells python to print plots inline with the notebook

rcParams['figure.figsize'] = 5,4
```

```python
x = range(1,10)
y= [1,2,3,4,0,4,3,2,1]

fig = plt.figure() #blank figure
ax = fig.add_axes([.1, .1, 1, 1])

ax.plot(x,y)
```

```python
x = range(1,10)
y= [1,2,3,4,0,4,3,2,1]

fig = plt.figure() #blank figure
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([1, 2, 4, 6, 8, 8.5, 9])
ax.set_yticks([i for i in range(5)])

ax.plot(x,y)
```

```python
x = range(1,10)
y= [1,2,3,4,0,4,3,2,1]

fig = plt.figure() #blank figure
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([1, 2, 4, 6, 8, 8.5, 9])
ax.set_yticks([i for i in range(5)])

ax.grid()
ax.plot(x,y)
```

```python
x = range(1,10)
y= [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)

plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
```

```python
z = [1, 2, 3, 4, 0.5]
color_theme = ['#A9A9A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
labels = ['A', 'B', 'C', 'D', 'E']

plt.pie(z, labels = labels, colors = color_theme)
plt.legend(labels, loc='best')

plt.show()
```

```python
x2 = range(1,6)
y2 = [10, 2, 3, 4, 0.5]

labels = ['A', 'B', 'C', 'D', 'E']

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xticks(range(1,6))
ax.set_xticklabels(labels, rotation = 60, fontsize = 'medium')

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')


ax.plot(x2, y2)
```

```python
#Annotating

x2 = range(1,6)
y2 = [10, 2, 3, 4, 0.5]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xticks(range(1,6))

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.annotate('Max Value', xy= (1,10), xytext=(2,10), arrowprops=dict(
    facecolor='black', shrink=0.05))

ax.plot(x2, y2)
```

# Statistical data graphics

Statistical plots allow viewers to identify *outliers*, *visualize distributions*, *deduce variable types*, and *discover relationships and core relations between variables* in a dataset.

**Histograms** are very simple plots that are used to show variable distribution.   
**Scatter plots** are used to show relationships between variables. Scatter plot matrices show core relations between variables.  

**Box plots** show variable spread and are useful for outlier detection.
